```python
def diffxsec(costheta,A_FB=-1):
    norm = 2.*((1.+1./3.))
    return ((1+costheta**2)+A_FB*costheta)/norm
```

```python
x = np.linspace(-1,1,100)
y = diffxsec(x)
```

```python
plt.plot(x,y)
```

```python
def rej_sample_costheta(nsamples = 100000):
    samples = []
    x = np.linspace(-1,1,num = 1000)
    maxval = np.max(diffxsec(x))
    while len(samples) < nsamples:
        xprop  = np.random.uniform(-1,1)
        ycut = np.random.random()
        yprop = diffxsec(xprop)/maxval
        if yprop/maxval < ycut:
            continue
        samples.append(xprop)
    return np.array(samples)
```

```python
v,b,p = plt.hist(rej_sample_costheta(), bins = 100, normed=True)
plt.plot(x,y)
```
